## Create a calculator with at least 15 math functions 


```python
def calculator(on):
    
    import math 

    while on==1:
        print()
        print("0. exit")
        print("1. addition")
        print("2. subtract")
        print("3. multiply")
        print("4. divide")
        print("5. exponents")
        print("6. areas")
        print("7. perimeters")
        print("8. quadratic formula")
        print("9. physics equations")
        
        choice=int(input("enter choice: "))

        if choice==1:
            print("Add")
            a=int(input("enter a- "))
            b=int(input("enter b- "))
            print(f"result: {a+b}")
        elif choice==2:
            print("Subtract")
            a=int(input("enter a- "))
            b=int(input("enter b- "))
            print(f"= {a-b}")
        elif choice==3:
            print("Multiply")
            a=int(input("enter a- "))
            b=int(input("enter b- "))
            print(f"= {a*b}")
        elif choice==4:
            print("Division a/b")
            a=int(input("enter a- "))
            b=int(input("enter b- "))
            print(f"= {a/b}")
        elif choice==5:
            print("Exponents a^b")
            a=int(input("enter base: "))
            b=int(input("enter power: "))
            print(f"{a}^{b}={a**b}")
        elif choice==6:
            print("Areas")
            print("1. rectangle")
            print("2. triangle")
            print("3. circle")
            jj=int(input("Enter choice: "))
            if jj==1:
                a=int(input("enter side a = "))
                b=int(input("enter side b = "))
                print(f"area of rectangle with sides {a} and {b} = {a*b}")
            elif jj==2:
                print("Area of triangle using 3 sides")
                a=None
                b=None
                c=None
                sides=[]
                a=int(input("enter side a = "))
                b=int(input("enter side b = "))
                c= int(input("enter side c = "))
                sp=(a+b+c)/2 # semi-perimeter
                sides.append(a)
                sides.append(b)
                sides.append(c)
                def calcArea(*args):
                    return math.sqrt(sp*((sp-a)*(sp-b)*(sp-c)))
                print(f"approximate triangle area = {calcArea(*sides)} square units")
            elif jj==3:
                print("Area of circle")
                from math import pi # or math.pi
                def area_circle(r):
                    return pi*(r**2)
                r=int(input("enter radius: "))
                print(f"the area of the circle is- {area_circle(r)}")
            else:
                print("invalid input, calculator shutting down.")
                break
        elif choice==7:
            print("Perimeters")
            print("1. rectangle")
            print("2. triangle")
            print("3. circle")
            pp=int(input("Enter choice: "))
            if pp==1:
                print("perimeter of rectangle:")
                a=int(input("enter side a = "))
                b=int(input("enter side b = "))
                print(f"perimeter = {(2*a)+(2*b)}")
            elif pp==2:
                print("perimeter of triangle:")
                import math
                a=None
                b=None
                c=None
                sides=[]
                a=int(input("enter side a = "))
                b=int(input("enter side b = "))
                c= int(input("enter side c = "))
                sp=(a+b+c)/2 
                sides.append(a)
                sides.append(b)
                sides.append(c)
                def calcPerimeter(*args):
                    return a+b+c
                print(f"perimeter = {calcPerimeter(*sides)} linear units")
            elif pp==3:
                print("perimeter of circle:")
                from math import pi
                def peri_circle(r):
                    return 2*pi*r
                r=int(input("enter radius: "))
                print(f"the perimeter of the circle is- {peri_circle(r)}")
            else:
                print("invalid input. calculator shutting down.")
                break
        elif choice==8:
            print("Quadratic formula")
            import math 
            def quadratic(a,b,c):
                x1=(-b+math.sqrt((b**2)-(4*a*c)))/(2*a)
                x2=(-b-math.sqrt((b**2)-(4*a*c)))/(2*a)
                return x1,x2
            a=int(input("enter a = "))
            b=int(input("enter b = "))
            c=int(input("enter c = "))
            print(f"x+,x- = {quadratic(a,b,c)}")
        
        elif choice==9:
            print("Choose Physics Equation:")
            print("1. final velocity")
            print("2. final distance")
            print("3. acceleration ")
            mm=int(input("eq : "))
            if mm==1:
                v=int(input("enter start V : "))
                a=int(input("acc : "))
                t=int(input("time : "))
                print(f"final velocity = {v+(a*t)}")
            elif mm==2:
                print("distance")
                x=int(input("enter start position : "))
                v=int(input("enter velocity : "))
                a=int(input("enter acc : "))
                t=int(input("enter time : "))
                print("final position = ", x+(v*t)+((1/2)*(a*(t**2))))
            elif mm==3:
                print("acceleration")
                v=int(input("enter start v : "))
                V=int(input("enter final V : "))
                t=int(input("enter time : "))
                print(f"acceleration = {(V-v)/t}")
        
        else:
            print("exit")
            break
```
